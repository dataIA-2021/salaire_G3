```python
def parse(url):
    html = requests.get(url)
    soup = BeautifulSoup(html.content, 'html.parser', from_encoding="utf-8")
    df = pd.DataFrame(columns=["Title","Location","Company","Salary", "Synopsis"])
    for jobs in soup.find_all(class_= "result" ):
        try: 
            title = title = jobs.find(class_='jobTitle').text.replace('\n', '')
        except:
            title = 'None'
        try:
            location = jobs.find(class_="companyLocation").text.replace('\n', '')
        except:
            location = 'None'
        try: 
            company = jobs.find(class_='companyName').text.replace('\n', '')
        except:
            company = 'None'
        try:
            salary = jobs.find(class_= 'salary-snippet').text
        except:
            salary = 'None'
        #synopsis = jobs.find(class_= 'job-snippet').text.text.replace('\n', '')
        df = df.append({'Title':title, 'Location':location, 'Company':company, 'Salary':salary}, ignore_index=True)
    return df
```

```python
parse(url)
```
