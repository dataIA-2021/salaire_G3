```python
def parse(url):
    html = requests.get(url)
    soup = BeautifulSoup(html.content, 'html.parser', from_encoding="utf-8")
    df = pd.DataFrame(columns=["Title","Location","Companie","Salary", "Résumé"])
    for jobs in soup.find_all(class_= "result" ):
        try: 
            title = jobs.find(class_='jobTitle').text.replace('\n', '')
        except:
            title = 'None'
        try:
            location = jobs.find(class_="companyLocation").text.replace('\n', '')
        except:
            location = 'None'
        try: 
            compagnie = jobs.find(class_='compagieName').text.replace('\n', '')
        except:
            compagnie = 'None'
        try:
            salary = jobs.find(class_= 'salary-snippet').text
        except:
            salary = 'None'
        try:
            resumé = jobs.find(class_= 'job-snippet').text.replace('\n', '')
        except:
            salary = 'None'
        try: jobkey = jobs.find(class_='jobTitle').text.replace('\n', '')
            
        df = df.append({'Title':title, 'Location':location, 'Company':compagnie, 'Salary':salary}, ignore_index=True)
    return df
```
